```python
import pandas as pd
import re
```

```python
df = pd.read_excel('../data/usa_festival_info.xlsx')
```

```python
df
```

```python
df_cleaned = df.astype(str).apply(lambda x: x.str.replace(r'\[\d+\]', '', regex=True))
```

```python
df_cleaned = df_cleaned.apply(lambda x: x.str.replace(r'\[[a-zA-Z]\]', '', regex=True))
```

```python
# df_cleaned = df_cleaned.apply(lambda x: x.str.replace(r'\[[a-zA-Z]\]', '', regex=True))
# df_cleaned = df_cleaned.apply(lambda x: x.str.replace(r' †\[n \d+\]', '', regex=True))
df_cleaned = df_cleaned.apply(lambda x: x.str.replace(r'\[n \d+\]', '', regex=True))
```

```python
df_cleaned.to_excel('../data/usa_festival_info.xlsx',index=False)
```
