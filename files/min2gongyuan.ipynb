```python
import pandas as pd
```

```python
df = pd.read_excel('D:/项目文件/交通出行/台湾/驾驶和交通安全/驾驶和交通安全.xlsx', sheet_name='Sheet1')
```

```python
def year_add(x):
    if pd.notna(x):  # 检查是否为非空值
        parts = x.split('-')
        if len(parts) == 3:  # 确保日期格式正确
            return f"{int(parts[0]) + 1911}-{parts[1]}-{parts[2]}"
    return x  # 如果为空值或格式错误，返回原值

df['发布时间'] = df['发布时间'].apply(year_add)
```
